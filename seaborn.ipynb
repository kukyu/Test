````python
import seaborn as sns

# 使用默认主题，如遇AttributeError请升级seaborn
sns.set_theme()

# 加载seaborn的样例数据集
tips = sns.load_dataset("tips", cache=True, data_home='seaborn_sample')

# 使用 relplot 创建散点图矩阵
sns.relplot(
    data=tips,
    x="total_bill", y="tip",    # x轴和y轴的变量
    col="time",                 # 按时间分割成两列
    hue="smoker",               # 根据吸烟者/非吸烟者着色
    style="smoker",             # 根据吸烟者/非吸烟者的样式
    size="size",                # 根据顾客数量调整点的大小
)
````

````python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
````

````python
# 显示了核密度估计（KDE）曲线。
sns.displot(data=tips, x="total_bill", col="time", kde=True)
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", size="size")
````

pairplot 是 Seaborn 中的一个函数，用于绘制数据集中多个变量之间的关系。它创建一个散点图矩阵，其中包含了数据集中每对变量之间的散点图和单变量的直方图或核密度估计曲线。
``` Python
sns.pairplot(data, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None,
             kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, dropna=True, plot_kws=None, diag_kws=None,
             grid_kws=None, size=None)
```
data: 数据集，可以是 DataFrame 或数组。

hue: 根据某个分类变量对数据进行着色，用于区分不同类别的数据。

vars: 选择要包含在图表中的变量。

kind: 图表类型，例如 "scatter"（散点图）、"reg"（回归图）、"kde"（核密度估计）等。

diag_kind: 对角线上的图表类型，可以是 "hist"（直方图）或 "kde"（核密度估计）。

palette: 用于着色的调色板。

````python
# 加载iris数据集
iris = sns.load_dataset('iris', cache=True, data_home='seaborn_sample')
# hue='species' 表示根据花的种类对数据进行着色
# diag_kind='kde' 表示对角线上显示核密度估计曲线
# kind='scatter' 表示其他位置上使用散点图
# palette='husl' 表示使用 "husl" 调色板进行着色
sns.pairplot(iris, hue='species', diag_kind='kde', kind='scatter', palette='husl')
````

````python
# 引入seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# 字体问题，默认中文无法显示
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'

# 数据集
x = ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']
y = [123, 145, 152, 182, 147, 138, 189, 201, 203, 211, 201, 182]

# 设置画布大小
plt.figure(figsize=(8, 6))
# 画一个柱形图
# plt.bar(x, y)
sns.barplot(x=x, y=y)
# 设置坐标轴范围
plt.ylim(50, 250)
# 添加网格线
plt.grid(axis='y', linestyle='--', linewidth=1, color='grey')
# 添加标题
plt.title("销售趋势图", fontdict={'family':'Microsoft YaHei', 'size': 15}, loc='left')
# 添加xy轴标题
plt.xlabel("月份", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 12}, labelpad=10)
plt.ylabel("销售额（万元）", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 12}, labelpad=10)

plt.show()
````
